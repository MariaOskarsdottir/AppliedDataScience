# Matplotlib tutorial

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.
Documentation: https://matplotlib.org/stable/index.html

```python
# Pandas for managing datasets
import numpy as np

# Matplotlib
from matplotlib import pyplot as plt
# To display your plots inside the notebook.
%matplotlib inline
```

## Basic plotting

```python
fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]);  # Plot some data on the axes.
```


## Coding styles

### The object-oriented and the pyplot interfaces

As noted above, there are essentially two ways to use Matplotlib:

- Explicitly create Figures and Axes, and call methods on them (the
  "object-oriented (OO) style").
- Rely on pyplot to automatically create and manage the Figures and Axes, and
  use pyplot functions for plotting.

So one can use the OO-style



```python
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend();  # Add a legend.
```

or the pyplot-style:



```python
x = np.linspace(0, 2, 100)  # Sample data.

plt.figure(figsize=(5, 2.7))
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend();
```

## Scatter plot
We make a simple scatter plot using the Pokmon data,

First we import hte datsaset.

```python
import pandas as pd
# Read dataset
df = pd.read_csv('Pokemon.csv')#, index_col=0)
# Display first 5 observations
df.head()
```

A basics scatter plot using the `scatter` method.

```python
# Making a simple scatter plot
fig, ax = plt.subplots()
ax.scatter(df['Attack'], df['Defense']) 
```

We make it again with costomized axes and a title.

```python
# Plot some data on the axes.
fig, ax = plt.subplots()
ax.scatter(df['Attack'], df['Defense']) 
ax.set_xlabel('Attack')  # Add an x-label to the axes.
ax.set_ylabel('Defense')  # Add a y-label to the axes.
ax.set_title("Scatter Plot")  # Add a title to the axes.
```

A different way to make the plot using Pandas.

```python
fig, ax = plt.subplots()
ax.scatter('Attack', 'Defense',data=df) # Use the data argument.
ax.set_xlabel('Attack')  # Add an x-label to the axes.
ax.set_ylabel('Defense')  # Add a y-label to the axes.
ax.set_title("Scatter Plot")  # Add a title to the axes.
```
